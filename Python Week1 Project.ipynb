```python
import pandas as pd
weather_data = pd.read_csv(r"C:\Users\User\OneDrive\Desktop\Data Analysis\Q\1. Weather Data.csv")
weather_data.head(10)
```

all records where the weather was exactly clear:

```python
clear_weather = weather_data[weather_data['Weather']=='Clear']
clear_weather.head(10)
```

Number of times the wind speed was exactly 4km/hr

```python
wind_speed_counts = weather_data[weather_data['Wind Speed_km/h'] == 4].shape[0]
wind_speed_counts
```

The mean visibility of the dataset

```python
mean_visibility = weather_data['Visibility_km'].mean()
mean_visibility
```

Number of records where the wind speed is greater than 24km/hr and visibility equal to 25km

```python
records = weather_data[(weather_data['Wind Speed_km/h']>24) & (weather_data['Visibility_km']==25)]
records.head(10)
```

Checking NULL values present in the dataset

```python
null_values = weather_data.isnull().sum()
null_values
```

Rename the column 'Weather' to 'Weather_Condition'

```python
weather_data.rename(columns = {'Weather':'Weather_Condition'}, inplace = True)
```

```python
weather_data.columns
```

The mean value of each column for each weather condition

```python
columns_mean = weather_data[['Temp_C','Dew Point Temp_C','Rel Hum_%','Wind Speed_km/h','Visibility_km','Press_kPa','Weather_Condition']]
grouped_mean = columns_mean.groupby(['Weather_Condition'], as_index = False).mean()
grouped_mean
```

all instances where the weather is clear and the relativee humidity is greater than 50, or visibility is above 40.

```python
filtered_data = weather_data[(weather_data['Weather_Condition']=='Clear') & (weather_data['Rel Hum_%']>50) | (weather_data['Visibility_km']>40)]
filtered_data.head(10)
```

the number of weather condition that include snow.

```python
snow_conditions = weather_data[weather_data['Weather_Condition'].str.contains('Snow', case=False, na=False)]
num_snow = snow_conditions['Weather_Condition'].nunique()
num_snow
```
